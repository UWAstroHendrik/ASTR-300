## Part I

$$ \Large {\displaystyle f(x)=3e^{-{\frac {x^{2}}{8}}}} \sin(x/3)$$

* Find the first four terms of the Taylor expansion of the above equation (at x = 0)
* Make a plot of the function
* Plot size 10 in x 4 in
* X limts -5, 5
* Y limits -2, 2
* On the same plot:
  * Over-plot the 1-term Taylor expansion using a different color/linetype/linewidth
  * Over-plot the 2-term Taylor expansion using a different color/linetype/linewidth
  * Over-plot the 3-term Taylor expansion using a different color/linetype/linewidth
  * Over-plot the 4-term Taylor expansion using a different color/linetype/linewidth

```python
my_np_x = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
x = sp.symbols('x')
```

```python
my_np_x1 = 3 * np.exp((-my_np_x**2)/8)*np.sin(my_np_x/3)
my_np_sp1 = 3 * sp.exp((-x**2)/8)*sp.sin(x/3)
```

```python
my_np_sp1
```

```python
my_taylor = sp.series(my_np_sp1, x, x0=0, n=8)

my_taylor.removeO().n(2)
```

```python
my_np_1term = my_np_x
my_np_2term = my_np_x - 0.14 * my_np_x**3
my_np_3term = my_np_x - 0.14 * my_np_x**3 + 0.01 * my_np_x**5
my_np_4term = my_np_x - 0.14 * my_np_x**3 + 0.01 * my_np_x**5 - 0.00048 * my_np_x**7
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(10,4)

fig.tight_layout()

ax.set_xlim(-5,5)
ax.set_ylim(-2,2)

ax.set_xlabel("X")
ax.set_ylabel("Y")

ax.plot(my_np_x, my_np_x1, color='r', marker='None', linestyle='-', linewidth=6)

ax.plot(my_np_x, my_np_1term, color='b', marker='None', linestyle='-', linewidth = 2, label="1-term")
ax.plot(my_np_x, my_np_2term, color='g', marker='None', linestyle='-', linewidth = 2, label="2-term")
ax.plot(my_np_x, my_np_3term, color='k', marker='None', linestyle='-', linewidth = 2, label="3-term")
ax.plot(my_np_x, my_np_4term, color='y', marker='None', linestyle='-', linewidth = 2, label="4-term")

ax.legend(loc = 0);
```
