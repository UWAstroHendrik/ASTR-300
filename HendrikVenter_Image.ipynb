## Create the following plots:

* They can be all in one plot, or as seperate plot (your choice).
* All images should have RA and DEC coords on the labels
* Put a title on each plot
  * **Plot 1**: Your image
  * **Plot 2**: Your image `(values < mean set to min)`
  * **Plot 3**: Your image `(values < (mean + 3 * std) set to min)`
  * **Plot 4**: Your image `(values < (mean + 7 * std) set to min)`
 

```python
data_file = "./Data/M42.fits"

fits.info(data_file)
```

```python
object_data   = fits.getdata(data_file, 0)
object_header = fits.getheader(data_file, 0)
```

```python
my_wcs = WCS(object_header)
```

```python
fig, ax = plt.subplots(1,7)
fig.set_size_inches(50,10)

fig.tight_layout(w_pad = 6)

ax[0].imshow(object_data, cmap=plt.cm.gist_heat, origin="lower")
ax[0].set_xlabel('Greens')

ax[1].imshow(object_data, cmap=plt.cm.bone, origin="lower")
ax[1].set_xlabel('cubehelix');

ax[2].imshow(object_data, cmap=plt.cm.cubehelix, origin="lower")
ax[2].set_xlabel('hot')

ax[3].imshow(object_data, cmap=plt.cm.CMRmap, origin="lower")
ax[3].set_xlabel('gist_heat')

ax[4].imshow(object_data, cmap=plt.cm.Reds, origin="lower")
ax[4].set_xlabel('plasma')

ax[5].imshow(object_data, cmap=plt.cm.Blues, origin="lower")
ax[5].set_xlabel('viridis')

ax[6].imshow(object_data, cmap=plt.cm.Greens, origin="lower")
ax[6].set_xlabel('RYB')


plt.savefig('M42 T.png')
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111, projection = my_wcs)

fig.set_size_inches(10,10)
fig.tight_layout(w_pad = 6)

ax1.set_title("M42", fontsize = 18)

ax1.set_xlabel("RA", fontsize = 18)
ax1.set_ylabel("Dec", fontsize = 18)
ax1.imshow(object_data, cmap=plt.cm.bone, origin="lower")

plt.savefig('M42 1.png')
```

```python
copy_data2 = np.copy(object_data)
```

```python
copy_data2[copy_data2 < copy_data2.mean()] = copy_data2.min()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111, projection = my_wcs)

fig.set_size_inches(10,10)
fig.tight_layout(w_pad = 6)

ax1.set_title("M42", fontsize = 18)
ax1.set_xlabel("RA", fontsize = 18)
ax1.set_ylabel("Dec", fontsize = 18)
ax1.imshow(copy_data2, cmap=plt.cm.bone, origin="lower")
plt.savefig('M42 2.png')
```

```python
copy_data3 = np.copy(object_data)
```

```python
copy_data3[copy_data3 < (copy_data3.mean() + 3 * copy_data3.std())] = copy_data3.min()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111, projection = my_wcs)

fig.set_size_inches(10,10)
fig.tight_layout(w_pad = 6)

ax1.set_title("M42", fontsize = 18)
ax1.set_xlabel("RA", fontsize = 18)
ax1.set_ylabel("Dec", fontsize = 18)
ax1.imshow(copy_data3, cmap=plt.cm.bone, origin="lower")
plt.savefig('M42 3.png')
```

```python
copy_data4 = np.copy(object_data)
```

```python
copy_data4[copy_data4 < (copy_data4.mean() + 7 * copy_data4.std())] = copy_data4.min()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111, projection = my_wcs)

fig.set_size_inches(10,10)
fig.tight_layout(w_pad = 6)

ax1.set_title("M42", fontsize = 18)
ax1.set_xlabel("RA", fontsize = 18)
ax1.set_ylabel("Dec", fontsize = 18)
ax1.imshow(copy_data4, cmap=plt.cm.bone, origin="lower")
plt.savefig('M42 4.png')
```
